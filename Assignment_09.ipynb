```python
import numpy as np
import pandas as pd
import datetime
import plotly.express as px
```

```python
df = px.data.tips()
df
```

```python
px.scatter(df, x="total_bill", y="tip", marginal_x="box", marginal_y="violin", color="sex", trendline="ols", facet_row="time", facet_col="day",
       category_orders={"day": ["Thur", "Fri", "Sat", "Sun"], "time": ["Lunch", "Dinner"]})
```

```python
fig = px.bar(df, x="size", y="total_bill", color="sex", facet_row="smoker")
fig.show()
```

```python
fig = px.scatter_matrix(df)
fig.show()
```
